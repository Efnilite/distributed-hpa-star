```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Set style for better-looking plots with colorblind-friendly palette
sns.set_style("whitegrid")
sns.set_palette("colorblind")  # Colorblind-friendly palette
plt.rcParams['figure.figsize'] = (14, 8)
plt.rcParams['font.size'] = 11
```

```python
fig, axes = plt.subplots(3, 1, figsize=(5, 15))

fig.suptitle("A* - Logarithmic environment comparison, average over 5 runs", fontsize=12, fontweight='bold')
# Data for each scenario
scenarios = ['scene_mp_2p_01: (260,180) to (1565,1745)', 'scene_mp_2p_04: (170,170) to (2000,2600)', 'maze: (1,1) to (999,999)']
cpu_times = [[0.089, 0.57, 0.57], [0.18, 0.91, 0.95], [0.001, 0.025, 0.024]]
memory_usage = [[10.69, 10.75, 10.75], [17.28, 17.30, 17.32], [4.52, 4.55, 4.54]]
heuristics = ['Manhattan', 'Chebyshev', 'Euclidean']

for idx, (scenario, cpu, mem) in enumerate(zip(scenarios, cpu_times, memory_usage)):
    ax = axes[idx]
    x = np.arange(len(heuristics))
    width = 0.35
    
    ax.bar(x - width/2, cpu, width, label='Path CPU time (s)', alpha=0.8)
    ax.bar(x + width/2, mem, width, label='Memory (MB)', alpha=0.8)
    
    ax.set_xlabel('Heuristic')
    ax.set_ylabel('Value')
    ax.set_title(scenario)
    ax.set_xticks(x)
    ax.set_xticklabels(heuristics)
    ax.set_yscale('log')
    ax.set_ylim([5e-4, 20])
    ax.legend()
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
